# DEFINING ALL MY FUNCTIONS: GLOROT INIT, FORWARD, BACKWARD PROP, AND UPDATE PARAM

```python
#STEP 1
#INITIALISING WEIGHTS USING GLOROT INITIALISIATION
```

```python
def initialize_parameters_uniform(input_size,h1_size,h2_size,h3_size,output_size):
  M1=np.sqrt(6/(input_size + h1_size))
  W1=np.random.uniform(-M1,M1,(input_size, h1_size))
  b1 = np.zeros((1, h1_size))

  M2 = np.sqrt(6 / (h1_size + h2_size))
  W2 = np.random.uniform(-M2, M2, (h1_size, h2_size))
  b2 = np.zeros((1, h2_size))

  M3 = np.sqrt(6 / (h2_size + h3_size))
  W3 = np.random.uniform(-M3, M3, (h2_size, h3_size))
  b3 = np.zeros((1, h3_size))

  M4 = np.sqrt(6 / (h3_size + output_size))
  W4 = np.random.uniform(-M4, M4, (h3_size, output_size))
  b4 = np.zeros((1, output_size))

  return W1, b1, W2, b2, W3, b3, W4, b4
```

```python
#STEP 2: DEFINING ACTIVATIONS AND CROSS ENTROPY LOSS
```

```python
def sigmoid(Z):
    return 1 / (1 + np.exp(-Z))

def sigmoid_derivative(Z):
    s = sigmoid(Z)
    return s * (1 - s)

def relu(Z):
    return np.maximum(0, Z)

def relu_derivative(Z):
    return np.where(Z > 0, 1, 0)

def tanh(Z):
    return np.tanh(Z)

def tanh_derivative(Z):
    return 1 - np.tanh(Z) ** 2

def softmax(Z):
    exp_Z = np.exp(Z - np.max(Z, axis=1, keepdims=True))
    return exp_Z / np.sum(exp_Z, axis=1, keepdims=True)

def cross_entropy_loss(y_true, y_pred):
    m = y_true.shape[0]
    log_likelihood = -np.log(y_pred[range(m), y_true.argmax(axis=1)])
    loss = np.sum(log_likelihood) / m
    return loss
```

```python
#STEP3: DEFINING A FORWARD PASS
```

```python
def forward_pass(X, W1, b1, W2, b2, W3, b3, W4, b4):
    Z1 = np.dot(X, W1) + b1
    A1 = sigmoid(Z1)

    Z2 = np.dot(A1, W2) + b2
    A2 = sigmoid(Z2)

    Z3 = np.dot(A2, W3) + b3
    A3 = sigmoid(Z3)

    Z4 = np.dot(A3, W4) + b4
    A4 = softmax(Z4)

    cache = (X, Z1, A1, Z2, A2, Z3, A3, Z4, A4)
    return A4, cache
```

```python
#STEP4: DEFINING BACKWARD PASS
```

```python
def backward_pass(X, y_true, cache, W4, W3, W2, W1):
    X, Z1, A1, Z2, A2, Z3, A3, Z4, A4 = cache
    m = X.shape[0]

    #o/p layer
    dZ4 = A4 - y_true
    dW4 = np.dot(A3.T, dZ4) / m
    db4 = np.sum(dZ4, axis=0, keepdims=True) / m

    #Layer3
    dA3 = np.dot(dZ4, W4.T)
    dZ3 = dA3 * sigmoid_derivative(Z3)
    dW3 = np.dot(A2.T, dZ3) / m
    db3 = np.sum(dZ3, axis=0, keepdims=True) / m

    #Layer2
    dA2 = np.dot(dZ3, W3.T)
    dZ2 = dA2 * sigmoid_derivative(Z2)
    dW2 = np.dot(A1.T, dZ2) / m
    db2 = np.sum(dZ2, axis=0, keepdims=True) / m

    #Layer1
    dA1 = np.dot(dZ2, W2.T)
    dZ1 = dA1 * sigmoid_derivative(Z1)
    dW1 = np.dot(X.T, dZ1) / m
    db1 = np.sum(dZ1, axis=0, keepdims=True) / m

    grads = {
        "dW4": dW4, "db4": db4,
        "dW3": dW3, "db3": db3,
        "dW2": dW2, "db2": db2,
        "dW1": dW1, "db1": db1
    }
    return grads
```

```python
#STEP5: UPDATE PARAM
```

```python
def update_parameters(params, grads, learning_rate):
    W1, b1, W2, b2, W3, b3, W4, b4 = params
    dW4, db4 = grads["dW4"], grads["db4"]
    dW3, db3 = grads["dW3"], grads["db3"]
    dW2, db2 = grads["dW2"], grads["db2"]
    dW1, db1 = grads["dW1"], grads["db1"]

    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1

    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2

    W3 -= learning_rate * dW3
    b3 -= learning_rate * db3

    W4 -= learning_rate * dW4
    b4 -= learning_rate * db4

    return W1, b1, W2, b2, W3, b3, W4, b4
```

```python
#STEP6: MINI BATCH
```

```python
def mini_batch_gradient_descent(X_train, y_train, X_val, y_val, input_size, h1_size, h2_size, h3_size, output_size, epochs, batch_size, learning_rate):
    W1, b1, W2, b2, W3, b3, W4, b4 = initialize_parameters_uniform(input_size, h1_size, h2_size, h3_size, output_size)
    num_samples = X_train.shape[0]

    train_losses = []
    val_losses = []
    val_accuracies = []

    for epoch in range(epochs):
        #Shuffle
        indices = np.arange(num_samples)
        np.random.shuffle(indices)
        X_train_shuffled = X_train[indices]
        y_train_shuffled = y_train[indices]

        #Process each mini-batch
        for start in range(0, num_samples, batch_size):
            end = min(start + batch_size, num_samples)
            X_batch = X_train_shuffled[start:end]
            y_batch = y_train_shuffled[start:end]

            #Forward pass
            A4, cache = forward_pass(X_batch, W1, b1, W2, b2, W3, b3, W4, b4)

            #loss
            loss = cross_entropy_loss(y_batch, A4)

            #Backward pass
            grads = backward_pass(X_batch, y_batch, cache, W4, W3, W2, W1)

            #Update param
            params = (W1, b1, W2, b2, W3, b3, W4, b4)
            W1, b1, W2, b2, W3, b3, W4, b4 = update_parameters(params, grads, learning_rate)

        #Val
        A4_val, _ = forward_pass(X_val, W1, b1, W2, b2, W3, b3, W4, b4)
        val_loss = cross_entropy_loss(y_val, A4_val)
        val_accuracy = np.mean(np.argmax(A4_val, axis=1) == np.argmax(y_val, axis=1))

        #metrics
        train_losses.append(loss)
        val_losses.append(val_loss)
        val_accuracies.append(val_accuracy)

        #Print Loss
        print(f"Epoch {epoch + 1}/{epochs}, Loss: {loss:.4f}, Val Loss: {val_loss:.4f}, Val Accuracy: {val_accuracy:.4f}")

    return W1, b1, W2, b2, W3, b3, W4, b4, train_losses, val_losses, val_accuracies

```

```python
#Evaluate on test set
def evaluate(X_test, y_test, W1, b1, W2, b2, W3, b3, W4, b4):
    A4, _ = forward_pass(X_test, W1, b1, W2, b2, W3, b3, W4, b4)
    predictions = np.argmax(A4, axis=1)
    true_labels = np.argmax(y_test, axis=1)
    accuracy = np.mean(predictions == true_labels)
    return accuracy
```

```python
#GEN Pred
def generate_predictions(X_test, W1, b1, W2, b2, W3, b3, W4, b4):
    A4, _ = forward_pass(X_test, W1, b1, W2, b2, W3, b3, W4, b4)
    predictions = np.argmax(A4, axis=1)
    return predictions
```

```python
def calculate_accuracy(X, y, W1, b1, W2, b2, W3, b3, W4, b4):
    predictions = generate_predictions(X, W1, b1, W2, b2, W3, b3, W4, b4)
    true_labels = np.argmax(y, axis=1)
    accuracy = np.mean(predictions == true_labels)
    return accuracy
```

```python
import torch
from torch.utils.data import DataLoader, TensorDataset
from torchvision import datasets, transforms

#Define transforms
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])

#Load MNIST dataset
train_dataset = datasets.MNIST(root='data', train=True, transform=transform, download=True)
test_dataset = datasets.MNIST(root='data', train=False, transform=transform, download=True)

#Split
train_size = int(0.8 * len(train_dataset))
val_size = len(train_dataset) - train_size
train_dataset, val_dataset = torch.utils.data.random_split(train_dataset, [train_size, val_size])

#Create DataLoaders
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

class MLP(nn.Module):
    def __init__(self):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(28*28, 500)
        self.fc2 = nn.Linear(500, 250)
        self.fc3 = nn.Linear(250, 100)
        self.fc4 = nn.Linear(100, 10)

    def forward(self, x):
        x = x.view(-1, 28*28)
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = torch.relu(self.fc3(x))
        x = self.fc4(x)
        return x

model = MLP()
```

```python
criterion = nn.CrossEntropyLoss()
criterion
```

```python
optimizer = optim.Adam(model.parameters(), lr=0.01)
optimizer
```

```python
def train(model, train_loader, criterion, optimizer, epochs=30):
    model.train()
    for epoch in range(epochs):
        running_loss = 0.0
        for images, labels in train_loader:
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()

        if (epoch+1) % 200 == 0:
            print(f'Epoch {epoch+1}, Loss: {running_loss/len(train_loader)}')
```

```python
def evaluate(model, test_loader):
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in test_loader:
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    accuracy = 100 * correct / total
    print(f'Accuracy of the network on the test images: {accuracy}%')
```

```python
train(model, train_loader, criterion, optimizer)
evaluate(model, test_loader)
```

```python
class MLP(nn.Module):
    def __init__(self):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(28*28, 500)
        self.fc2 = nn.Linear(500, 250)
        self.fc3 = nn.Linear(250, 100)
        self.fc4 = nn.Linear(100, 10)

    def forward(self, x):
        x = x.view(-1, 28*28)
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = torch.relu(self.fc3(x))
        x = self.fc4(x)
        return x
```

```python
from sklearn.metrics import accuracy_score

def train_and_evaluate(model, train_loader, test_loader, weight_decay):
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.01, weight_decay=weight_decay)  # Use weight_decay

    model.train()
    for epoch in range(15):
        running_loss = 0.0
        for images, labels in train_loader:
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()

    model.eval()
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for images, labels in test_loader:
            outputs = model(images)
            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    accuracy = accuracy_score(all_labels, all_preds) * 100
    return accuracy
```

```python
alpha = [0.0001, 0.001, 0.01, 0.1, 10]
accuracies = {}

for weight_decay in alpha:
    model = MLP()
    accuracy = train_and_evaluate(model, train_loader, test_loader, weight_decay=weight_decay)
    accuracies[weight_decay] = accuracy
    print(f'Weight Decay: {weight_decay}, Test Accuracy: {accuracy:.2f}%')

plt.figure(figsize=(10, 6))
plt.plot(alpha, [accuracies[wd] for wd in alpha], marker='o')
plt.xscale('log')
plt.xlabel('alpha')
plt.ylabel('Test Accuracy (%)')
plt.title('Test Accuracy vs Weight Decay')
plt.grid(True)
plt.show()
```

```python
def forward_pass(X, W1, b1, W2, b2, W3, b3, W4, b4):
    Z1 = np.dot(X, W1) + b1
    A1 = sigmoid(Z1)

    Z2 = np.dot(A1, W2) + b2
    A2 = tanh(Z2)

    Z3 = np.dot(A2, W3) + b3
    A3 = relu(Z3)

    Z4 = np.dot(A3, W4) + b4
    A4 = softmax(Z4)

    cache = (X, Z1, A1, Z2, A2, Z3, A3, Z4, A4)
    return A4, cache
```

```python
def backward_pass(X, y_true, cache, W4, W3, W2, W1):
    X, Z1, A1, Z2, A2, Z3, A3, Z4, A4 = cache
    m = X.shape[0]

    #o/p layer
    dZ4 = A4 - y_true
    dW4 = np.dot(A3.T, dZ4) / m
    db4 = np.sum(dZ4, axis=0, keepdims=True) / m

    #Layer3
    dA3 = np.dot(dZ4, W4.T)
    dZ3 = dA3 * sigmoid_derivative(Z3)
    dW3 = np.dot(A2.T, dZ3) / m
    db3 = np.sum(dZ3, axis=0, keepdims=True) / m

    #Layer2
    dA2 = np.dot(dZ3, W3.T)
    dZ2 = dA2 * tanh_derivative(Z2)
    dW2 = np.dot(A1.T, dZ2) / m
    db2 = np.sum(dZ2, axis=0, keepdims=True) / m

    #Layer1
    dA1 = np.dot(dZ2, W2.T)
    dZ1 = dA1 * relu_derivative(Z1)
    dW1 = np.dot(X.T, dZ1) / m
    db1 = np.sum(dZ1, axis=0, keepdims=True) / m

    grads = {
        "dW4": dW4, "db4": db4,
        "dW3": dW3, "db3": db3,
        "dW2": dW2, "db2": db2,
        "dW1": dW1, "db1": db1
    }
    return grads
```
